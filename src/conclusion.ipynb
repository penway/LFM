```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget

legends = []
fid_hist = []
LFM_min = []
DC_min = []

# for p in [1, 0.5, 0.1, 0.02]:
#     for type in ["LFM", "DC"]:
for p in [1, 0.5, 0.1, 0.02]:
    for type in ["LFM", "DC"]:

        log_path = "E:\\Caldron\\LFM\\result\\celeb\\{}_{}\\novel.csv".format(type, p)
        legends.append("{}_{}".format(type, p))
        res_df = pd.read_csv(log_path)

        fid_hist.append(res_df[" fid"])

        if type == "LFM":
            LFM_min.append(min(res_df[" fid"].tolist()))
        else:
            DC_min.append(min(res_df[" fid"].tolist()))
```

```python
plt.figure()
for i, hist in enumerate(fid_hist):
    plt.plot(range(0, 30001, 500), hist)
plt.legend(legends)
plt.yscale("log")
plt.show()
```

```python
plt.figure()
plt.plot(['full dataset', '50%', '10%', '5%'], LFM_min)
plt.plot(['full dataset', '50%', '10%', '5%'], DC_min)
print(LFM_min)
print(DC_min)
plt.legend(["LFM", "DC"])
plt.show()
```
